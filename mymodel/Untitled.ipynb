```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("test_640.jpg")
plt.imshow(img)
```

```python
kernel = np.ones((5,5),np.uint8)
img_t = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
plt.imshow(img_t)
```

```python
cv2.imwrite('test_640_2.jpg', img_t)
```
